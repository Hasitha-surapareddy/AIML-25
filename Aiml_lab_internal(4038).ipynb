```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('/content/IRIS.csv')
df.head()
```

```python
df.describe()
```

```python
df.isna()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df.hist(figsize=(10, 8))
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']] = scaled_features
```

```python
print(scaled_features)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
knn_model_2 = KNeighborsClassifier(n_neighbors=2)
knn_model_2.fit(X_train, y_train)
```

```python
knn_model_3 = KNeighborsClassifier(n_neighbors=3)
knn_model_3.fit(X_train, y_train)
```

```python

y_pred_2 = knn_model_2.predict(X_test)
accuracy_2 = accuracy_score(y_test, y_pred_2)
print(f"Accuracy with 2 neighbors: {accuracy_2}")

y_pred_3 = knn_model_3.predict(X_test)
accuracy_3 = accuracy_score(y_test, y_pred_3)
print(f"Accuracy with 3 neighbors: {accuracy_3}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
```

```python

y_train_pred_2 = knn_model_2.predict(X_train)


y_test_pred_2 = knn_model_2.predict(X_test)

y_train_pred_3 = knn_model_3.predict(X_train)


y_test_pred_3 = knn_model_3.predict(X_test)
```

```python

def evaluate_model(y_true, y_pred, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted')
    conf_matrix = confusion_matrix(y_true, y_pred)

    print(f"--- {model_name} ---")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print("Confusion Matrix:")
    print(conf_matrix)
    print("\n")
```

```python

evaluate_model(y_train, y_train_pred_2, "KNN (k=2) - Training Data")
evaluate_model(y_train, y_train_pred_3, "KNN (k=3) - Training Data")


evaluate_model(y_test, y_test_pred_2, "KNN (k=2) - Testing Data")
evaluate_model(y_test, y_test_pred_3, "KNN (k=3) - Testing Data")
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
error_rate = []
for i in range(1, 40):
         knn = KNeighborsClassifier(n_neighbors=i)
         knn.fit(X_train, y_train)
         pred_i = knn.predict(X_test)
         error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed', marker='o',
markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```
